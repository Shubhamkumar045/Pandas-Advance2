```python
Q1. Write a code to print the data present in the second row of the dataframe, df.
Answer--import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Print the data in the second row (index 1)
second_row_data = df.iloc[1]
print("Data in the second row:")
print(second_row_data)

Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Answer--
A)loc (label-based indexing):
Syntax: df.loc[row_label, column_label]
Usage: It is used for selection by label. You provide the row label(s) and column label(s) to get the desired data.
Example:# Accessing a specific value
value = df.loc[0, 'column_name']

# Accessing a specific row
row_data = df.loc[0]

# Accessing a specific column
column_data = df.loc[:, 'column_name']

B)iloc (integer-location based indexing):
Syntax: df.iloc[row_index, column_index]
Usage: It is used for selection by integer index. You provide the row index(es) and column index(es) to get the desired data.
Example:# Accessing a specific value
value = df.iloc[0, 1]

# Accessing a specific row
row_data = df.iloc[0]

# Accessing a specific column
column_data = df.iloc[:, 1]

Key Differences:
loc uses labels, while iloc uses integer indices.
In loc, the end index is inclusive, whereas in iloc, it is exclusive.
Both can be used for single label/index, lists, slices, boolean masks, etc.
loc is more explicit about what you are asking, using labels, while iloc is more implicit, using integer positions.

Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].
Answer--import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a DataFrame
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Reindexing using the specified order
reindex_order = [3, 0, 1, 2]
new_df = df1.reindex(reindex_order)

# Accessing row with label '2' using loc
loc_result = new_df.loc[2]

# Accessing row at index 2 using iloc
iloc_result = new_df.iloc[2]

# Printing the results
print("Using new_df.loc[2]:")
print(loc_result)
print("\nUsing new_df.iloc[2]:")
print(iloc_result)

Q4. Write a code to find the following statistical measurements for the above dataframe df1:
(i) mean of each and every column present in the dataframe.
(ii) standard deviation of column, ‘column_2’
Answer--import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a DataFrame
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# (i) Mean of each and every column
column_means = df1.mean()

# (ii) Standard deviation of 'column_2'
std_column_2 = df1['column_2'].std()

# Printing the results
print("(i) Mean of each and every column:")
print(column_means)

print("\n(ii) Standard deviation of 'column_2':")
print(std_column_2)


(df1.mean() calculates the mean of each column.
df1['column_2'].std() calculates the standard deviation of 'column_2')

Q5. Replace the data present in the second row of column, ‘column_2’ by a string variable then find the
mean of column, column_2.
If you are getting errors in executing it then explain why.
[Hint: To replace the data use df1.loc[] and equate this to string data of your choice.]
Answer--You can replace the data in the second row of 'column_2' with a string variable
and then calculate the mean of 'column_2'.


import pandas as pd
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a DataFrame
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Replacing data in the second row of 'column_2' with a string variable
df1.loc[2, 'column_2'] = 'some_string'

# Attempting to calculate the mean of 'column_2'
try:
    # This line will raise an error if there are mixed data types
    mean_column_2 = df1['column_2'].mean()
    print("Mean of 'column_2':", mean_column_2)
except Exception as e:
    print(f"Error: {e}")

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
Answer--window functions are operations applied to a specified window of data points within a series or DataFrame. 
These functions allow you to perform calculations over a sliding or expanding window of data, making them useful
for tasks such as time-series analysis, signal processing, and rolling statistics. Window functions
are particularly valuable when dealing with ordered data.

Here are some common types of window functions in pandas:

A)Rolling Functions:

Description: These functions operate on a rolling (moving) window of a specified size over the data.
They are useful for calculating moving averages, cumulative sums, and other rolling statistics
# Calculate the rolling mean with a window size of 3
df['rolling_mean'] = df['column_name'].rolling(window=3).mean()

B)Expanding Functions:
Description: These functions operate on an expanding window that starts from the beginning 
of the data and grows with each data point. They are useful for calculating cumulative statistics.
Example:# Calculate the expanding sum
df['expanding_sum'] = df['column_name'].expanding().sum()

C)Exponential Moving Average (EMA):
Description: EMA is a type of rolling window function that gives more weight to recent
data points, making it responsive to changes in the data.
Example:# Calculate the exponential moving average with a specified span
df['ema'] = df['column_name'].ewm(span=10, adjust=False).mean()

D)Window Aggregations:

Description: These functions perform aggregations over a specified window. Examples include window sum, mean, min, max, etc.
Example:# Calculate the window sum with a window size of 5
df['window_sum'] = df['column_name'].rolling(window=5).sum()

E)Custom Window Functions:

Description: You can define and apply custom window functions using the apply method in combination with a custom function.
Example:# Define a custom function
def custom_window_function(window_data):
    # Perform custom calculations
    return result

# Apply the custom function to a rolling window
df['custom_result'] = df['column_name'].rolling(window=3).apply(custom_window_function)

Q7. Write a code to print only the current month and year at the time of answering this question.
[Hint: Use pandas.datetime function]
Answer--the pd.to_datetime function along with the pd.Timestamp.now() function to get the
current date and time, and then extract the month and year.

import pandas as pd

# Get the current date and time
current_datetime = pd.Timestamp.now()

# Extract the current month and year
current_month = current_datetime.month
current_year = current_datetime.year

# Print the result
print(f"Current Month: {current_month}")
print(f"Current Year: {current_year}")

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.
Answer--import pandas as pd

def calculate_time_difference():
    # Get input from the user for the two dates
    date1_str = input("Enter the first date (YYYY-MM-DD): ")
    date2_str = input("Enter the second date (YYYY-MM-DD): ")

    try:
        # Convert input strings to pandas datetime objects
        date1 = pd.to_datetime(date1_str)
        date2 = pd.to_datetime(date2_str)

        # Calculate the time difference
        time_difference = date2 - date1

        # Extract days, hours, and minutes
        days = time_difference.days
        hours, remainder = divmod(time_difference.seconds, 3600)
        minutes, _ = divmod(remainder, 60)

        # Display the result
        print(f"\nTime Difference: {days} days, {hours} hours, {minutes} minutes")

    except ValueError as e:
        print(f"Error: {e}")

# Call the function to calculate and display the time difference
calculate_time_difference()

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.
Answer--
```
